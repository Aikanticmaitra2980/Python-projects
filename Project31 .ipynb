```python
import numpy as np
Students_data={}
```

```python
def display_students():
    print("\n" + "-"*40)
    print("STUDENT DETAILS")
    print("-"*40)

    for roll, data in Students_data.items():
        print(f"\nRoll Number : {roll}")
        print(f"Name        : {data['Name']}")
        print("Marks:")
        for subject, mark in data['subject_marks'].items():
            print(f"  {subject.strip():<15} : {mark}")
```

```python
Students_data = {}

def students(num_students):
    Subjects = input("Enter Subjects (comma separated): ").split(',')

    for i in range(num_students):
        while True:
            try:
                Roll_number = int(input("Enter Roll_no: "))
                break
            except ValueError:
                print("Invalid input for Roll_number. Please enter an integer.")

        Name = input("Enter Name: ")

        marks = []
        for sub in Subjects:
            while True:
                try:
                    mark_input = input(f"Enter marks for {sub.strip()}: ")
                    marks.append(int(mark_input))
                    break
                except ValueError:
                    print("Invalid input for marks. Please enter an integer.")

        subject_marks = dict(zip(Subjects, marks))

        Students_data[Roll_number] = {
            'Name': Name,
            'Subjects': Subjects,
            'subject_marks': subject_marks
        }

    display_students()
```

```python
def student_stat():
  for roll_no,student_info in Students_data.items():
    average_marks=np.average(list(student_info['subject_marks'].values()))
    Maximum=np.max(list(student_info['subject_marks'].values()))
    Minimum=np.min(list(student_info['subject_marks'].values()))
    print("average marks for Each student is",average_marks)
    print("Maximum marks for Each student is",Maximum)
    print("Minimum marks for Each student is",Minimum)
```

```python
def remove_student():
  roll_no=int(input("Enter Roll_no:"))
  if roll_no in Students_data:
    del Students_data[roll_no]
    print("Student removed successfully")
```

```python
def Add_student():
  while True:
    try:
      Roll_number=int(input("Enter Roll_no:"))
      if Roll_number in Students_data:
        print("Roll number already exists. Please enter a unique Roll_no.")
      else:
        break
    except ValueError:
      print("Invalid input for Roll_no. Please enter an integer.")

  Name=input("Enter Name:")
  Subjects=input("Enter Subjects:").split(',')
  Students_data[Roll_number]={'Name':Name,'Subjects':Subjects}

  subject_marks = {}
  for sub in Subjects:
    while True:
      try:
        mark_input = input(f"Enter marks for {sub}: ")
        subject_marks[sub] = int(mark_input)
        break
      except ValueError:
        print("Invalid input for marks. Please enter an integer.")
  Students_data[Roll_number]['subject_marks']=subject_marks
  print("Student added successfully")
```

```python
commands:{'Resultsheet',
            'Show',
            'Remove',
            'Add',
            'Stats'}
```

```python
M=int(input("Enter number of students:"))
decision="Yes"
while decision=="Yes":
  commands=input("Enter commands:")
  if commands=='Resultsheet':
   students(M)
  elif commands=='Show':
   display_students()
  elif commands=='Remove':
   remove_student()
  elif commands=='Add':
   Add_student()
  elif commands=='Stats':
   student_stat()
  elif commands=='Exit':
   break
  else:
   print("Invalid command.")
  decision = input("Do you want to continue? (Yes/No):")
```
