```python
import numpy as np
```

```python
Students=[]
Marks=[]
student_dict=dict(zip(Students,Marks))
```

```python

def studentsf(n):
    for i in range(n):
        temp1=input("Enter Name of the student")
        Students.append(temp1)
```

```python
def marks(m):
    for i in range(m):
        temp2=int(input("Enter marks"))
        Marks.append(temp2)
```

```python
def stats_student(Marks):
    avg=np.average(Marks)
    maximum=np.max(Marks)
    minimum=np.min(Marks)
    print("Average marks of this class is",avg)
    print("Highest marks of this class is",maximum)
    print("Minimum marks of this class is",minimum)
```

```python
def update_dict():
    global student_dict
    student_dict=dict(zip(Students,Marks))
```

```python
def remove_student():
    update_dict()
    rstudent = input("Enter name to remove: ")
    if rstudent in student_dict:
        idx = Students.index(rstudent)
        Students.pop(idx)
        Marks.pop(idx)
        update_dict()
        print("Updated list:", student_dict)
    else:
        print("Student not found.")
```

```python
def add_student():
    name=input("Enter new Student name:")
    mark=int(input("Enter marks:"))
    Students.append(name)
    Marks.append(mark)
    update_dict()
    print("Student added Succesfully!")
    print(student_dict)
```
